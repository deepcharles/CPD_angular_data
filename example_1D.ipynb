```python
import numpy as np
import matplotlib.pyplot as plt
from utils import pelt, apart

pi = np.pi
np.random.seed(123)
```

```python
# create 1D angular signal
signal = np.concatenate([np.random.normal(m, 0.3, 5) for m in [0, 2, 4]])
signal = signal.reshape((-1, 1))

# shift the signal into [0, 2pi)
signal[signal < 0]    += 2*pi
signal[signal > 2*pi] -= 2*pi

# plot signal
plt.figure(figsize=(6, 2))
plt.scatter(np.arange(len(signal)), signal, label='signal', color='black')
plt.legend(loc='best')
plt.xlabel("time step")
plt.ylabel("angle (radian)")
plt.show()
```

```python
# discrete means set
Theta = np.linspace(0, 2*pi, 6, endpoint = False).reshape(-1, 1)

# Plot Theta
plt.figure(figsize=(6, 2))
plt.scatter(range(len(signal)), signal, label='signal', color='black')
for i, theta in enumerate(Theta):
    plt.axhline(y = theta, color='blue', linestyle='--', label='thetas' if i == 0 else "")

plt.legend(loc='best')
plt.xlabel('time step')
plt.ylabel('angle (radian)')
plt.show()
```

```python
# optimal partitioning
lda = 0.5
chpnts, signal_mean = pelt(signal, Theta, lda)

# plotting solution
plt.figure(figsize=(6, 2))
plt.scatter(range(len(signal)), signal, label='signal', color='black')
plt.plot(signal_mean, label='signal mean', color='red', alpha=0.5)
for i, cp in enumerate(chpnts):
    plt.axvline(x = cp + 0.5, color='black', linestyle='--', label='changepoints' if i == 0 else "")
plt.legend(loc='best')
plt.xlabel('time step')
plt.ylabel('angle (radian)')
plt.show()
```

```python
# optimal partitioning
lda = 0.5
chpnts, signal_mean = apart(signal, Theta, lda)

# plotting solution
plt.figure(figsize=(6, 2))
plt.scatter(range(len(signal)), signal, label='signal', color='black')
plt.plot(signal_mean, label='signal mean', color='red', alpha=0.5)
for i, cp in enumerate(chpnts):
    plt.axvline(x = cp + 0.5, color='black', linestyle='--', label='changepoints' if i == 0 else "")
plt.legend(loc='best')
plt.xlabel('time step')
plt.ylabel('angle (radian)')
plt.show()
```
